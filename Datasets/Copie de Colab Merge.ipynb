# Merge Datasets Transport PMR

```python
df_arrets_access = pd.read_csv("/content/drive/MyDrive/Restolympiques/Datasets pré Nettoyage/df_arrets_access.csv", sep = ";")
df_arrets_lignes = pd.read_csv("/content/drive/MyDrive/Restolympiques/Datasets pré Nettoyage/df_arrets-lignes.csv", sep = ";")
df_emplacement_gares = pd.read_csv("/content/drive/MyDrive/Restolympiques/Datasets pré Nettoyage/emplacement_gares.csv", sep = ";")
```

```python
print(df_arrets_access.head().to_markdown())
```

```python
df_arrets_access.shape
```

```python
print(df_arrets_lignes.head().to_markdown())
```

```python
df_arrets_lignes.shape
```

```python
df_merge_arrets = pd.merge(df_arrets_access, df_arrets_lignes, on = "stop_id", how = "inner")
print(df_merge_arrets.head().to_markdown())
```

```python
df_merge_arrets = df_merge_arrets.drop(columns = ["route_id_y", "route_long_name_y", "stop_name_y", "OperatorName_y", "Pointgeo_y"])
```

```python
df_merge_arrets.shape
```

```python
print(df_emplacement_gares.head().to_markdown())
```

```python
df_merge_arrets_gares = pd.merge(df_merge_arrets, df_emplacement_gares, left_on = "Pointgeo_x", right_on = "Geo Point", how = "left")
```

```python
print(df_merge_arrets_gares.head().to_markdown())
```

```python
df_merge_arrets_gares.shape
```

```python
df_merge_arrets_gares.columns
```

```python
df_merge_arrets_gares = df_merge_arrets_gares.drop(columns = ['latitude_y', 'longitude_y'])
```

```python
df_merge_arrets_gares.rename(columns = {"route_id_x" : "route_id", "route_long_name_x": "route_long_name",
                                        "stop_name_x": "stop_name", "OperatorName_x": "OperatorName",
                                        "Pointgeo_x": "Pointgeo", "Nom_commune_x": "Nom_commune",
                                        "latitude_x": "latitude", "longitude_x": "longitude"})
```

```python
print(df_merge_arrets_gares.head().to_markdown())
```

```python
df_merge_arrets_gares["Pointgeo_x"].drop_duplicates()
```

```python
print(df_merge_arrets_gares.head().to_markdown())
```

```python
df_merge_arrets_gares.rename(columns = {"route_id_x" : "route_id", "route_long_name_x": "route_long_name",
                                        "stop_name_x": "stop_name", "OperatorName_x": "OperatorName",
                                        "Pointgeo_x": "Pointgeo", "Nom_commune_x": "Nom_commune",
                                        "latitude_x": "latitude", "longitude_x": "longitude",
                                        "Nom_commune_y": "Nom_commune"})
```
