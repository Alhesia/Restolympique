```python
print(df_trippy.head().to_markdown())
```

```python
df_trippy = df_trippy.drop(columns = ["default_language", "reviews_count_in_default_language"])
```

```python
#df_trippy = df_trippy.drop(columns = ["vegetarian_friendly", "vegan_options", "gluten_free"])
```

```python
df_trippy = df_trippy.drop(columns = ["atmosphere", "keywords"])
```

```python
print(df_trippy.head().to_markdown())
```

```python
df_trippy = df_trippy.drop(columns = ["claimed", "awards", "popularity_detailed", "price_range", "open_days_per_week", "open_hours_per_week", "working_shifts_per_week", "popularity_generic"])
```

```python
df_trippy['features'] = df_trippy['features'].str.replace(',', '')
```

```python
df_trippy["meals"] = df_trippy["meals"].str.replace(",", "")
df_trippy["cuisines"] = df_trippy["cuisines"].str.replace(",", "")
```

```python
print(df_trippy.head().to_markdown())
```

```python
def conca(restaurant_link, restaurant_name):
  url_prefix = "https://www.tripadvisor.fr/Restaurant_Review-"
  return f"{url_prefix}{restaurant_link}-{restaurant_name}-Paris_Ile_de_France.html"

df_trippy["complete_url"] = df_trippy.apply(lambda row: conca(row["restaurant_link"], row["restaurant_name"]), axis=1)
```

```python
print(df_trippy.head().to_markdown())
```

```python
df_trippy.shape
```

```python
df_trippy["special_diets"].value_counts()
```
